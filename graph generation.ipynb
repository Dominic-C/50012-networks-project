```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# this function takes in the path to a log file. Instructions to generate the log file is found on the readme in github
def get_bandwidth(path):
    with open(path, "r") as f:
        data = f.readlines()
    speed = []
    for i in range(4, len(data)):
        current_speed = data[i].split()[-1].strip("M\n")
        speed.append(float(current_speed))
        
    time = [i for i in range(len(speed))]
    return time, speed
```

```python
time, speed = get_bandwidth("sample.log")
```

```python
def basic_plot(x_axis, y_axis, title, x_label, y_label):
    plt.figure(figsize=[10,6])
    plt.plot(x_axis, y_axis)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
basic_plot(time,speed,"Bandwidth vs Time", "time(s)", "bandwidth MB/s")
```
