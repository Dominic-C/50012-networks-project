```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
http = []
http3 = []
def get_timing(path):
    lines = []
    with open(path, "r") as f:
        for line in f:
            lines.append(line.split())
    f.close()
    timing = lines[-1][8].split(":")
#     print(timing[-2])
    return int(timing[-1])+int(timing[-2])*60
#     return timing
# get_timing("http3/testhttp3.3.log")
```

```python
for i in range(1,11):
    time = get_timing("http/test."+str(i)+".log")
    http.append(time)
    
for i in range(1,11):
    time = get_timing("http3/testhttp3."+str(i)+".log")
    http3.append(time)
```

```python
# basic_plot(time,speed,"Iteration vs Time", "Iterations", "Time(s)")
plt.plot(http,'r',label = "HTTP 2")
plt.plot(http3,'g',label = "HTTP 3")
plt.xlabel("Iterations")
plt.ylabel("Time(s)")
plt.legend()
plt.savefig('http2vshttp3.png')
```
